## Scratch - Pandas Dataframe demo

In this notebook, we demonstrate how to work with data using Pandas, and visualize it using Matplotlib and Seaborn. This notebook might lead to unwished reactions for users with a cat allergy.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Say hello to Annika with some cat fun! üò∏
print("Hey Annika! üéâ Time for some cat-tastic data visualization! üêæ Let's see which cat breed reigns supreme!")

# Cat data: Name of breeds and their "popularity"
data = {
    'Cat Breed': ['Persian', 'Maine Coon', 'Bengal', 'Siamese', 'Ragdoll'],
    'Popularity': [85, 95, 90, 75, 88]
}

df = pd.DataFrame(data)

# Plotting with Seaborn (because cats deserve the best!)
plt.figure(figsize=(8, 6))
sns.barplot(x='Cat Breed', y='Popularity', data=df, palette='coolwarm')
plt.title('Cat Breeds Popularity (Based on Feline Fandom üòª)')
plt.xlabel('Cat Breed')
plt.ylabel('Popularity')
plt.show()

# Ending with more cat love üê±
print("And that's how cats rule the world of data visualization! Go ahead, Annika, and become a cat data expert! üêæüíª")
```
